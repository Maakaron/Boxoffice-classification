```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import matplotlib.patches as mpatches
from matplotlib import style
import seaborn as sns
style.use('ggplot')
data = pd.read_csv('movies-checkpoint_FINAL.csv')
pd.set_option('display.max_columns', None)
data = data.rename(columns={'Real GDP (trillions $)': 'GDP'})
del data['Unnamed: 0']
del data['usa_gross_income']
del data['worldwide_gross_income']

data
#Using Pearson Correlation
pt.figure(figsize=(30,30))
cor = data.corr()
k= sns.heatmap(cor, annot=True, cmap=pt.cm.Reds)
pt.show()

```

```python
#Correlation with output variable
cor_target = abs(cor["Classification"])
#Selecting highly correlated features
High_degree_cor = cor_target[cor_target>0.50]
High_degree_cor
Mid_degree_cor = cor_target[cor_target>0.30]
Mid_degree_cor
k=cor_target.drop(labels = 'Classification')
k=k.sort_values(axis=0, ascending=True)
# low_degree_cor = cor_target[cor_target<0.30]
# low_degree_cor
ax = k.plot(kind='bar', title ="Feature importance - correlation", figsize=(15, 10), legend=True, fontsize=12)
k
```

```python
genres = data.groupby('Classification')['Action','Adventure','Animation','Biography','Comedy','Crime','Drama','Family','Fantasy','History',
                                        'Horror','Music','Musical',	'Mystery','Romance','Sci-Fi','Sport','Thriller','War','Western'].sum()

display(genres)
#genres[['Action','Adventure','Animation','Biography','Comedy','Crime','Drama','Family','Fantasy','History',
#                                        'Horror','Music','Musical',	'Mystery','Romance','Sci-Fi','Sport','Thriller','War','Western']].plot(kind='bar')
genres.T.plot.bar(figsize=(20,10))
pt.show()
```

```python
GDP = data.Classification.value_counts().reset_index(name='GDP')

display(GDP)
```

```python
GDP1 = data.loc[data['Classification']==1,'GDP'].value_counts()
#del GDP1['level_0']

GDP1 = GDP1.to_frame()
```

```python
GDP1.rename(columns={ GDP1.columns[0]: "Occurances" }, inplace = True)

GDP1.columns
```

```python
GDP0 = data.loc[data['Classification']==0,'GDP'].value_counts()
#del GDP0['level_0']
GDP0 = GDP0.to_frame()
GDP0.rename(columns={ GDP0.columns[0]: "Occurances" }, inplace = True)
GDP0
```

```python
GDP_FIN = pd.merge(GDP1, GDP0, on='index', how='outer')

GDP_FIN = GDP_FIN.rename(columns={'GDP_x': '1', 'GDP_y': '0', 'index': 'GDP [Trillion $]'})

GDP_FIN = GDP_FIN.sort_values(by=['GDP [Trillion $]'], ascending=False)
GDP_FIN['MOVIES'] = GDP_FIN['1'] + GDP_FIN['0']
GDP_FIN['Success_procentage'] = GDP_FIN['1'] / GDP_FIN['MOVIES']
GDP_FIN
```

```python
GDP_FIN.plot(x='GDP [Trillion $]', y='Success_procentage', label='Movie success rate %')


```
